```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,criterion='absolute_error',max_features=5,random_state=0)
regressor=regressor.fit(X_train,Y_train)
```

```python
Y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
regressor.feature_importances_
```

```python
import pickle
fileName="final_RandomForest_1_Model.sav"
pickle.dump(regressor,open(fileName,'wb'))
```

```python
load_model=pickle.load(open("final_RandomForest_1_Model.sav",'rb'))
```
